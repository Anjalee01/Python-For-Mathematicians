```python
import matplotlib.pyplot as plt

# Sample data for three datasets
data1 = [7, 8, 5, 6, 9, 8, 6, 5, 4, 7, 6, 8, 5, 7]
data2 = [4, 5, 3, 5, 4, 6, 5, 7, 6, 5, 6, 4, 5]
data3 = [9, 10, 8, 9, 11, 9, 10, 8, 9, 10, 11]

# Combine data into a list
data = [data1, data2, data3]

# Create a box plot with custom colors
box = plt.boxplot(
    data,
    patch_artist=True,  # Enables color filling
    labels=['Dataset 1', 'Dataset 2', 'Dataset 3']
)

# Customizing colors
colors = ['lightblue', 'green', 'red']
for patch, color in zip(box['boxes'], colors):        # iterates over the boxes and assign the colors of boxes
    patch.set_facecolor(color)

# Add title and labels
plt.title('Box Plot with Multiple Datasets and Colors')
plt.xlabel('Datasets')
plt.ylabel('Values')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
data = [[7, 8, 5, 6, 9], [4, 5, 3, 5, 4], [4,5,9, 10, 8, 9, 11]]

# Horizontal box plot
plt.boxplot(data, patch_artist=True, vert=False, labels=['Group 1', 'Group 2', 'Group 3'])

# Customizing colors
colors = ['skyblue', 'lightpink', 'lightgray']
for patch, color in zip(plt.boxplot(data, patch_artist=True, vert=False)['boxes'], colors):
    patch.set_facecolor(color)

# Add title and labels
plt.title('Horizontal Box Plot')
plt.xlabel('Values')
plt.ylabel('Groups')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data with outliers
data = [7, 8, 5, 6, 50, 9, 8, 6, 7]

# Create a box plot
box = plt.boxplot(data, patch_artist=True, flierprops=dict(markerfacecolor='red', marker='o'))

# Customizing box color
box['boxes'][0].set_facecolor('lightblue')

# Add title and labels
plt.title('Box Plot with Highlighted Outliers')
plt.xlabel('Dataset')
plt.ylabel('Values')

plt.show()
```


#Histogram



```python
import matplotlib.pyplot as plt

# Sample data
data = [7, 8, 5, 6, 8,9, 6, 5, 4, 7, 6, 8, 5, 7,9,9,9]

# Create histogram
plt.hist(data, bins=5, color='lightblue', edgecolor='black')

# Add title and labels
plt.title('Basic Histogram')
plt.xlabel('Value Ranges')
plt.ylabel('Frequency')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Sample data
data = [7, 8, 5, 6, 9, 8, 6, 5, 4, 7, 6, 8, 5, 7,6,9,10]

# Histogram with customized density curve color
sns.histplot(
    data,
    kde=True,  bins = 7,
    color='green',  # Color for the bars
    edgecolor='black',  # Edge color for the bars
)

# Add title and labels
plt.title('Histogram with Custom KDE Curve Color')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
data = [7, 8, 5, 6, 9, 8, 6, 5, 4, 7, 6, 8, 5, 7]

# Create horizontal histogram
plt.hist(data, bins=5, orientation='horizontal', color='lightgreen', edgecolor='black')

# Add title and labels
plt.title('Horizontal Histogram')
plt.xlabel('Frequency')
plt.ylabel('Value Ranges')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for two groups
group1 = np.random.normal(5, 1, 100)       #100= random number, 1 = S.D, 5 = mean
group2 = np.random.normal(7, 1.5, 100)

# Create histograms
plt.hist(group1, bins=10, alpha=0.5, label='Group 1', color='blue', edgecolor='black')
plt.hist(group2, bins=10, alpha=0.5, label='Group 2', color='orange', edgecolor='black')

# Add title, labels, and legend
plt.title('Grouped Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = [7, 8, 5, 6, 9, 8, 6, 5, 4, 7, 6, 8, 5, 7]

# Styled histogram using Seaborn
sns.histplot(data, bins=3, kde=True, color='red')

# Add title and labels
plt.title('Seaborn Styled Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.show()
```
