# What is Pandas?
Pandas is a powerful and widely-used open-source data manipulation and analysis library in Python. It provides data structures and functions needed to efficiently handle and analyze large datasets. Pandas is particularly well-suited for working with structured data (i.e., data that can be organized into tables, like databases or spreadsheets).

It provides two primary data structures:

**Series:** A one-dimensional labeled array, similar to a list or a column in a table.
                                                                   
**DataFrame:** A two-dimensional table of labeled data, similar to a spreadsheet or SQL table, consisting of rows and columns.

# **How to retrieve a row?**
Pandas provide a unique method to retrieve rows from a Data frame. DataFrame.loc[] method is a method that takes only index labels and returns row or dataframe

```python
#row selection
newdata = {'Name':['Tom', 'nick', 'krish', 'jack'],
        'Age':[20, 21, 19, 18],
           "Grades":["a","b","c","d"]
        }
# frame1=pd.DataFrame(newdata)
labels=[f"Row_{i}" for i in range(1,5)]
frame1=pd.DataFrame(newdata,labels)
print("original dataset")
print(frame1)
print("----------------------------------------")
print("Dataset after applying loc method")
fr1=frame1.loc[['Row_1','Row_2']]
print(fr1)
```

```python
#iloc retrieve row based on postion or index
newdata = {'Name':['Tom', 'nick', 'krish', 'jack'],
        'Age':[20, 21, 19, 18],
           "Grades":["a","b","c","d"]
        }
labels=[f"Row_{i}" for i in range(1,5)]
frame1=pd.DataFrame(newdata,labels)
fr1=frame1.iloc[[1,2]]
print(fr1)
```

```python
#drop a row
newframe = {'Name':['Tom', 'nick', 'krish', 'jack'],
        'Age':[20, 21, 19, 18],
           "Grades":["a","b","c","d"]
        }
labels1=[f"Row_{i}" for i in range(1,5)]
frame2=pd.DataFrame(newframe,labels1)
frame3=frame2.drop(["Row_1","Row_2"], axis=0, inplace=True)
print("Original DataFrame")
print(frame2)

print("DataFrame after inplace=False")
print(frame3)
```

```python
import pandas as pd
# Original data dictionary with 5 rows
data_dict1 = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 45],
    'Occupation': ['Engineer', 'Doctor', 'Artist', 'Lawyer', 'Teacher'],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
    'Salary': [70000, 80000, 60000, 90000, 75000]
}

# Create the DataFrame from the dictionary
df = pd.DataFrame(data_dict1)

# Function to take input for new rows
def add_new_row():
    name = input("Enter Name: ")
    age = int(input("Enter Age: "))
    occupation = input("Enter Occupation: ")
    city = input("Enter City: ")
    salary = int(input("Enter Salary: "))

    return {'Name': name, 'Age': age, 'Occupation': occupation, 'City': city, 'Salary': salary} #returning a dictionary

# Create a list to store new rows
new_rows = []

# Loop to add multiple rows (ask user for each)
num_rows = int(input("How many rows would you like to add? "))
for i in range(num_rows):
    print(f"\nEnter details for row {i+1}:")
    new_row = add_new_row()
    print(new_row)
    new_rows.append(new_row)
    print(new_rows)

# Convert the list of new rows to a DataFrame
new_rows_df = pd.DataFrame(new_rows)
print(new_rows_df)
# Concatenate the original DataFrame with the new rows
df = pd.concat([df.iloc[:2], new_rows_df, df.iloc[2:]], ignore_index=True)

# Print the updated DataFrame
print("\nUpdated DataFrame:")
print(df)
```
